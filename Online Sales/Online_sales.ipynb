```python
#loading libraries needed
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the dataset
df = pd.read_csv("Online Sales Data.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
#checking for missing values
sns.heatmap(df.isnull(), yticklabels = "False", cbar = False, cmap = 'viridis')
```

```python
#checking the highest payment method
sns.set_style('darkgrid')
sns.histplot(x = 'Payment Method', data = df)
```

```python
#checking the region with the most sales
sns.set_style('darkgrid')
sns.histplot(x = 'Region', data = df)
```

```python
#Highest product
plt.figure(figsize=(10,5))
sns.set_style('darkgrid')
sns.histplot(x = "Product Category", data = df)
```

```python
#Total revenue from each product category
plt.figure(figsize = (10,5))
sns.set_style('whitegrid')
sns.barplot(x = "Product Category",y ="Total Revenue", data = df)
```

```python
#Total revenue from each region
plt.figure(figsize = (10,5))
sns.set_style('whitegrid')
sns.barplot(x = "Region",y ="Total Revenue", data = df)
```

```python
#Total revenue from each country based on Product category
plt.figure(figsize = (10,5))
sns.set_style('whitegrid')
sns.barplot(x = "Region",y ="Total Revenue",hue = "Product Category", data = df, )
```
